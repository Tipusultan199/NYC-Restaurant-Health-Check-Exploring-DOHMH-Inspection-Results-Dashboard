```python
import pandas as pd
import numpy as np
```

```python
df= pd.read_csv('DOHMH_New_York_City_Restaurant_Inspection_Results.csv')
df = df.reset_index(drop=True)
df.head(10)
```

```python
df.columns
```

```python
missing_values = df.isnull().sum()
missing_values
```

```python
df = df.dropna(subset=['DBA', 'BUILDING', 'ZIPCODE', 'CUISINE DESCRIPTION','ACTION','VIOLATION CODE','VIOLATION DESCRIPTION','INSPECTION TYPE','NTA','BBL','BIN','Council District','Latitude','Longitude','Census Tract','Community Board','SCORE'])
```

```python
df = df.drop('Location Point1', axis=1)
df = df.drop('PHONE', axis=1)
df = df.drop('GRADE DATE', axis=1)
```

```python
missing_values = df.isnull().sum()
missing_values
```

```python
df.shape
```

```python
from datetime import datetime, timedelta
# convert 'INSPECTION DATE' column to datetime format
df['INSPECTION DATE'] = pd.to_datetime(df['INSPECTION DATE'])

# calculate the date five years ago from today
five_years_ago = datetime.now() - timedelta(days=5*365)

# filter the dataset for inspections done in the last five years
df = df[df['INSPECTION DATE'] > five_years_ago]
```

```python
df
```

```python
# Fill missing values in the DBA and CUISINE DESCRIPTION columns with "Not Specified"
df = df.fillna({ 
                 'GRADE': 'Not Graded'
               })

```

```python
df
```

```python
missing_values = df.isnull().sum()
missing_values
```

```python
# Convert 'INSPECTION DATE' column to datetime format
df['INSPECTION DATE'] = pd.to_datetime(df['INSPECTION DATE'])

# Extract year from 'INSPECTION DATE' column and create new 'INSPECTION YEAR' column
df['INSPECTION YEAR'] = df['INSPECTION DATE'].dt.year
```

```python
df
```

```python
missing_values = df.isnull().sum()
missing_values
```

```python
# assuming the dataframe is named `df`
export_path = "final.csv"

# export the dataframe to CSV file
df.to_csv(export_path, index=False)
```
